Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new  column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.  Monday, Tuesday) corresponding to each date in the 'Date' column. 
For example, if df contains the following values: 

```
         Date 
0  2023-01-01 
1  2023-01-02 
2  2023-01-03 
3  2023-01-04 
4  2023-01-05
```
Your function should create the following DataFrame: 
```
        Date    Weekday 
0  2023-01-01   Sunday 
1  2023-01-02   Monday 
2  2023-01-03   Tuesday 
3  2023-01-04   Wednesday 
4  2023-01-05   Thursday 
```
The function should return the modified DataFrame.


````python
from  datetime import datetime

df =pd.DataFrame({'Date': [ "2023-01-01","2023-01-02","2023-01-03","2023-01-04","2023-01-05"]})
def weekday(df) :
    date_l= []
    for i in range(0,len(df)):    
        date = datetime.strptime(df['Date'][i], "%Y-%m-%d").strftime("%A")
        date_l.append(date)
    df['Weekday'] = date_l
    return df  

weekday(df)
````

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python  function to select all rows where the date is between '2023-01-01' and '2023-01-31'. 

````python
import pandas as pd
df =  pd.read_csv('sample sales data at csv.csv')
df['Date-Time'] = pd.to_datetime(df['Date-Time'], format='%d-%m-%Y %H:%M') # using this function converting dates ad date time format so that i can select dates

start_date = '2023-03-01 00:00:00'
end_date = '2023-03-31 23:59:59'

selecet_date = df.loc[(df['Date-Time'] >= start_date) & (df['Date-Time'] <= end_date)]
selecet_date
````
